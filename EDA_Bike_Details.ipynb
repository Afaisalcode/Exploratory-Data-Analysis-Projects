Observation:
 - ex_showroom_price contains 435 missing values (~41% of the dataset).
Since this column is important for resale value analysis, median imputation is used.

```python
# Handling Missing Values

df['ex_showroom_price'] = df['ex_showroom_price'].fillna(
    df['ex_showroom_price'].median()
)
```

```python
#Q9. What is the total number of bikes listed by each seller type?
df['seller_type'].value_counts()

#Visualization
sns.countplot(x='seller_type',
              data=df)

plt.show()
```

```python
#Q10.What is the relationship between selling_price and km_driven for first-owner bikes?
first_owner = df[df['owner'] == '1st owner']

#Correlation
first_owner[['selling_price',
             'km_driven']].corr()

#Scatter Plot

plt.figure(figsize=(8,5))

sns.scatterplot(
    x='km_driven',
    y='selling_price',
    data=first_owner
)

plt.title("Selling Price vs KM Driven (1st Owner)")
plt.show()


```

Interpretation
- The analysis shows a negative correlation coefficient of -0.24 between selling_price and km_driven for first-owner bikes. This confirms that as a motorcycle is driven more, its market resale value predictably decreases.”

```python
#Q11.Identify and remove outliers in the km_driven column using the IQR method.

#Calculate Quartiles
Q1 = df['km_driven'].quantile(0.25)

Q3 = df['km_driven'].quantile(0.75)

IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)

#Calculate Bounds
lower_bound = Q1 - 1.5 * IQR

upper_bound = Q3 + 1.5 * IQR

print(lower_bound)
print(upper_bound)

#Finding Outliers

outliers = df[
    (df['km_driven'] < lower_bound) |
    (df['km_driven'] > upper_bound)
]

print("Number of Outliers:",
      outliers.shape[0])

#Removing Outliers

df_clean = df[
    (df['km_driven'] >= lower_bound) &
    (df['km_driven'] <= upper_bound)
]

print(df_clean.shape)
```

```python
#Q12.Perform a bivariate analysis to visualize the relationship between year and selling_price.
plt.figure(figsize=(8,5))

sns.barplot(
    x='year',
    y='selling_price',
    data=df
)

plt.title("Year vs Selling Price")
plt.show()


#Regression Plot

plt.figure(figsize=(8,5))

sns.regplot(
    x='year',
    y='selling_price',
    data=df
)

plt.show()
```

Interpretation
- Upward trend → newer bikes have higher prices.
- Downward trend → older bikes have lower prices.

```python
#Q13.What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)?

current_year = 2026

df['age'] = current_year - df['year']

#Average Selling Price by Age
depreciation = df.groupby('age')['selling_price'].mean()

print(depreciation)

#Plot
depreciation.plot(
    kind='line',
    figsize=(10,5)
)

plt.title("Average Selling Price by Age")
plt.show()
```

```python
#14. Which bike names are priced significantly above the average price for their manufacturing year?

#Calculate Mean and Std

year_stats = df.groupby('year')['selling_price'].agg(
    ['mean','std']
)

year_stats.head()

#Merge with Original Dataset

df2 = df.merge(
    year_stats,
    on='year'
)

#Find High-Priced Bikes
high_price_bikes = df2[
    df2['selling_price']
    >
    (df2['mean'] + 2*df2['std'])
]

high_price_bikes[
    ['name',
     'year',
     'selling_price']
]


```

```python
#Q15.Develop a correlation matrix for numeric columns and visualize it using a heatmap.

#Select Numeric Columns

numeric_df = df.select_dtypes(
    include=np.number
)


#Correlation Matrix

corr_matrix = numeric_df.corr()

corr_matrix


#Heatmap
plt.figure(figsize=(10,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("Correlation Heatmap")
plt.show()
```
