### Dataset Description:
This dataset contains product listings, pricing details, customer ratings, and user reviews from Amazon. It provides insights into category-wise sales, discount strategies, consumer preferences, and feedback metrics.

#### Key Features:
- **product_id:** Unique identifier for each product.
- **product_name:** Name/title of the product.
- **category:** Product category path separated by '|'.
- **discounted_price:** Current selling price of the product on Amazon.
- **actual_price:** Retail price/MSRP of the product before discount.
- **discount_percentage:** Percentage of discount offered.
- **rating:** Average customer rating of the product.
- **rating_count:** Total number of customers who rated the product.
- **about_product:** Brief description or bullet points about the product features.
- **user_id / user_name:** Encrypted IDs and names of users who wrote reviews.
- **review_id / review_title / review_content:** Details regarding the text reviews posted by customers.

##Problems Identified:

- Data Types: Numeric columns like prices, discounts, and ratings are stored as text (object).

 - Garbage Characters: Pricing features contain currency symbols (₹) and commas (,), and discounts contain %.

 - Anomalous Values: The rating column has a corrupt row containing a text symbol (|), and rating_count has missing values.

##Solution Strategy:

- Cleaned all currency, percentage, and comma symbols using string replacement.

- Replaced the corrupt character (|) with NaN and handled missing records using column medians and zeros.

- Converted all relevant features into clean numeric formats (float64) and extracted the high-level main_category for analysis.

```python
# 1. Clean Prices & Discounts
for col in ['discounted_price', 'actual_price']:
    df[col] = df[col].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float) / 100

# 2. Clean Ratings & Handled Missing Values
df['rating'] = pd.to_numeric(df['rating'].astype(str).str.replace('[^0-9.]', '', regex=True), errors='coerce')
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)

df['rating_count'] = df['rating_count'].fillna(0)
df['rating'] = df['rating'].fillna(df['rating'].median())

# 3. Extract Main Category
df['main_category'] = df['category'].astype(str).str.split('|').str[0].str.strip()

# Check Final Output
df.info()
```

```python
#Q1. ## Q1. What is the average rating for each product category?
# Groupby lagakar har category ka average rating nikalenge
avg_rating_cat = df.groupby('main_category')['rating'].mean().sort_values(ascending=False)

print("Average Rating by Category:")
print(round(avg_rating_cat, 2))

# Ekdum simple bar chart
plt.figure(figsize=(10, 4))
plt.bar(avg_rating_cat.index, avg_rating_cat.values, color='skyblue', edgecolor='black')
plt.title('Average Rating by Product Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Q2. ## Q2. What are the top rating_count products by category?

# Har category me sabse zyada review (rating_count) wale top products
print("Top Product by Rating Count in each Category:")

# idxmax() se hume sabse badi value ka index mil jata hai
top_products = df.loc[df.groupby('main_category')['rating_count'].idxmax()]

# Sirf kaam ke 3 columns print karenge taaki readable lage
display(top_products[['main_category', 'product_name', 'rating_count']])
```

```python
## Q3. What is the distribution of discounted prices vs. actual prices?

# Dono prices ka average dekhne ke liye simple statistics
print("Summary of Prices:")
print(df[['actual_price', 'discounted_price']].describe())

# Ekdum simple subplots dono ka distribution dekhne ke liye
plt.figure(figsize=(12, 5))

# Actual Price ka histogram
plt.subplot(1, 2, 1)
plt.hist(df['actual_price'], bins=15, color='red', edgecolor='black', alpha=0.7)
plt.title('Distribution of Actual Prices')
plt.xlabel('Price (₹)')
plt.ylabel('Count')

# Discounted Price ka histogram
plt.subplot(1, 2, 2)
plt.hist(df['discounted_price'], bins=15, color='green', edgecolor='black', alpha=0.7)
plt.title('Distribution of Discounted Prices')
plt.xlabel('Price (₹)')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
## Q4. How does the average discount percentage vary across categories?

# Groupby se discount percentage ka average nikalenge
avg_discount_cat = df.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False)

print("Average Discount Percentage by Category:")
print(round(avg_discount_cat, 2))

# Simple horizontal bar plot
plt.figure(figsize=(10, 4))
plt.barh(avg_discount_cat.index, avg_discount_cat.values, color='orange', edgecolor='black')
plt.title('Average Discount Percentage across Categories')
plt.xlabel('Average Discount (%)')
plt.ylabel('Category')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
## Q5. What are the most popular product names?

# Kaunsa product name dataset me sabse zyada baar aaya hai (value_counts)
popular_products = df['product_name'].value_counts().head(5)

print("Top 5 Most Popular Product Names (by frequency in dataset):")
print(popular_products)

# Simple bar plot
plt.figure(figsize=(10, 4))
plt.barh(popular_products.index.str.slice(0, 40), popular_products.values, color='teal')
plt.title('Top 5 Most Frequent Product Names')
plt.xlabel('Count')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
## Q6. What are the most popular product keywords?

# Product names me se basic keywords nikalne ka sabse simple tarika
all_words = ' '.join(df['product_name']).lower().split()

# Ek simple pandas Series banakar words ka count nikalenge
word_counts = pd.Series(all_words).value_counts()

# Faltu ke chote words (like with, for, and) ko ignore karke kaam ke top keywords dikhayenge
print("Most Popular Keywords in Product Names:")
print(word_counts.head(10))
```

```python
## Q7. What are the most popular product reviews?

# Short reviews (review_title) me sabse common titles kaunse hain
popular_reviews = df['review_title'].value_counts().head(5)

print("Top 5 Most Common Review Titles:")
print(popular_reviews)
```

```python
## Q8. What is the correlation between discounted_price and rating?

# Pearson correlation nikalenge dono continuous features ke beech
price_rating_corr = df['discounted_price'].corr(df['rating'])
print("Correlation between Discounted Price and Rating:", round(price_rating_corr, 4))

# Simple scatter plot dono ka relation dekhne ke liye
plt.figure(figsize=(7, 4))
plt.scatter(df['discounted_price'], df['rating'], color='purple', alpha=0.5)
plt.title('Discounted Price vs Product Rating')
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Rating')
plt.show()
```

```python
## Q9. What are the Top 5 categories based on the highest ratings?

# Q1 me nikale gae data se top 5 categories filter karenge
top_5_rated_categories = df.groupby('main_category')['rating'].mean().sort_values(ascending=False).head(5)

print("Top 5 Categories with Highest Average Ratings:")
print(round(top_5_rated_categories, 2))
```
