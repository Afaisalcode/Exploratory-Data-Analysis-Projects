### Observation:
- The dataset is extremely clean.
- Only **Customer Name** has **1 missing value**.
- There are no other missing values across any of the remaining 15 columns.
- We will fill this single missing value with **'Unknown'** and proceed with removing duplicate rows (if any) to ensure perfect data integrity.

```python

df['Customer Name'] = df['Customer Name'].fillna('Unknown')

# Re-check missing values to confirm 0 nulls
df.isnull().sum()
```

```python
# Plot the distribution of car selling prices
plt.figure(figsize=(8, 5))
sns.histplot(df['Price ($)'], bins=30, kde=True, color='blue')
plt.title('Distribution of Car Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of customer annual income
plt.figure(figsize=(8, 5))
sns.histplot(df['Annual Income'], bins=30, kde=True, color='green')
plt.title('Distribution of Customer Annual Income')
plt.xlabel('Annual Income ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the count of cars based on transmission types
plt.figure(figsize=(6, 4))
sns.countplot(x='Transmission', data=df, palette='Set2')
plt.title('Count of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.show()
```

```python
#Q1. What is the average selling price of cars for each dealer, and how does it compare across different dealers?

# Step 1: Calculate the average price for each dealer
dealer_avg = df.groupby('Dealer_Name')['Price ($)'].mean().reset_index()

# Step 2: Sort the dealers by price from highest to lowest
dealer_avg = dealer_avg.sort_values(by='Price ($)', ascending=False)

print("--- Average Car Price by Dealer ---")
print(dealer_avg)

# Step 3: Plot a simple bar chart to compare the dealers
plt.figure(figsize=(10, 5))
sns.barplot(data=dealer_avg, x='Price ($)', y='Dealer_Name', palette='Blues_r')
plt.title('Average Car Selling Price by Dealer')
plt.xlabel('Average Price ($)')
plt.ylabel('Dealer Name')
plt.tight_layout()
plt.show()
```

```python
#Q2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
#trends?

# Step 1: Calculate the standard deviation (variation) of prices for each brand
brand_std = df.groupby('Company')['Price ($)'].std().reset_index()

# Step 2: Rename columns for simplicity
brand_std.columns = ['Company', 'Price_Variation']

# Step 3: Sort from highest variation to lowest
brand_std = brand_std.sort_values(by='Price_Variation', ascending=False)

print("--- Car Price Variation by Brand ---")
print(brand_std)

# Step 4: Plot a bar chart to visualize the variation
plt.figure(figsize=(12, 6))
sns.barplot(data=brand_std, x='Company', y='Price_Variation', palette='Reds_r')
plt.title('Car Price Variation (Standard Deviation) by Brand')
plt.xlabel('Car Brand (Company)')
plt.ylabel('Price Variation ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Q3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?

# Create a box plot to show price distribution across different transmission types
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Transmission', y='Price ($)', palette='Set2')
plt.title('Car Price Distribution by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
#Q4. What is the distribution of car prices across different regions?

# Create a box plot to see the price distribution across dealer regions
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Dealer_Region', y='Price ($)', palette='Set3')
plt.title('Car Price Distribution Across Different Regions')
plt.xlabel('Dealer Region')
plt.ylabel('Price ($)')
plt.show()
```

```python
#Q5. What is the distribution of cars based on body styles?

# Count how many cars are sold for each body style
body_style_counts = df['Body Style'].value_counts()
print("--- Car Count by Body Style ---")
print(body_style_counts)

# Create a simple pie chart to show the proportion of each body style
plt.figure(figsize=(8, 8))
plt.pie(body_style_counts, labels=body_style_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Distribution of Cars Based on Body Styles')
plt.show()
```

```python
#Q6. How does the average selling price of cars vary by customer gender and annual income?

# Step 1: Create a basic scatter plot colored by gender to check how income and price relate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income', y='Price ($)', hue='Gender', alpha=0.6, palette='Set1')
plt.title('Car Price vs Annual Income by Gender')
plt.xlabel('Annual Income ($)')
plt.ylabel('Price ($)')
plt.show()

# Step 2: Calculate and print the exact average car price for each gender
gender_avg_price = df.groupby('Gender')['Price ($)'].mean().reset_index()
print("--- Average Car Price by Customer Gender ---")
print(gender_avg_price)
```

```python
#Q7. What is the distribution of car prices by region, and how does the number of cars sold vary by region?

# Step 1: Count of cars sold per region
region_counts = df['Dealer_Region'].value_counts().reset_index()
region_counts.columns = ['Dealer_Region', 'Cars_Sold']

print("--- Cars Sold by Region ---")
print(region_counts)

# Step 2: Show both Price Distribution and Sales Volume in a subplot
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Left plot: Violin plot for Price Distribution by Region
sns.violinplot(ax=axes[0], data=df, x='Dealer_Region', y='Price ($)', palette='muted')
axes[0].set_title('Car Price Distribution by Region')
axes[0].set_xlabel('Dealer Region')
axes[0].set_ylabel('Price ($)')

# Right plot: Bar plot for Sales Volume by Region
sns.barplot(ax=axes[1], data=region_counts, x='Dealer_Region', y='Cars_Sold', palette='Blues_r')
axes[1].set_title('Number of Cars Sold by Region')
axes[1].set_xlabel('Dealer Region')
axes[1].set_ylabel('Number of Cars Sold')

plt.tight_layout()
plt.show()
```

```python
#Q8. How does the average car price differ between cars with different engine sizes?

# Step 1: Calculate the average price for each engine type
engine_prices = df.groupby('Engine')['Price ($)'].mean().reset_index()

# Step 2: Sort from highest average price to lowest
engine_prices = engine_prices.sort_values(by='Price ($)', ascending=False)

print("--- Average Price by Engine Size ---")
print(engine_prices)

# Step 3: Plot the comparison using a simple bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=engine_prices, x='Engine', y='Price ($)', palette='viridis')
plt.title('Average Car Price by Engine Type')
plt.xlabel('Engine Type')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Q9. How do car prices vary based on the customer’s annual income bracket?

# Step 1: Create a basic function to categorize annual income into brackets
def get_income_bracket(income):
    if income < 50000:
        return 'Low (<$50K)'
    elif income <= 100000:
        return 'Medium ($50K - $100K)'
    else:
        return 'High (>$100K)'

# Step 2: Apply the function to create a new column
df['Income_Bracket'] = df['Annual Income'].apply(get_income_bracket)

# Step 3: Calculate average car price for each income bracket
bracket_prices = df.groupby('Income_Bracket')['Price ($)'].mean().reset_index()
print("--- Average Car Price by Income Bracket ---")
print(bracket_prices)

# Step 4: Visualize using a box plot to see the full price range variation
plt.figure(figsize=(10, 5))
order_list = ['Low (<$50K)', 'Medium ($50K - $100K)', 'High (>$100K)']
sns.boxplot(data=df, x='Income_Bracket', y='Price ($)', order=order_list, palette='coolwarm')
plt.title('Car Price Distribution by Customer Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Car Price ($)')
plt.show()
```

```python
#Q10. What are the top 5 car models with the highest number of sales, and how does their price distribution look?

# Step 1: Identify the top 5 car models by sales count
top_models = df['Model'].value_counts().head(5).index.tolist()
print("Top 5 Sold Models:", top_models)

# Step 2: Filter the main dataframe to keep only these top 5 models
df_top_models = df[df['Model'].isin(top_models)]

# Step 3: Create a box plot to show price distribution for these top 5 models
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_top_models, x='Model', y='Price ($)', palette='Set1')
plt.title('Price Distribution of Top 5 Most Popular Car Models')
plt.xlabel('Car Model')
plt.ylabel('Price ($)')
plt.show()
```

```python
#Q11. How does car price vary with engine size across different car colors, and which colors have the highest price variation?

# Step 1: Calculate price standard deviation (variation) for each color to see which has the highest variation
color_variation = df.groupby('Color')['Price ($)'].std().reset_index()
color_variation.columns = ['Color', 'Price_Variation']
color_variation = color_variation.sort_values(by='Price_Variation', ascending=False)

print("--- Car Price Variation by Color ---")
print(color_variation)

# Step 2: Plot how price changes across different Engine types and Colors (top 5 popular colors for clarity)
top_colors = df['Color'].value_counts().head(5).index.tolist()
df_colors = df[df['Color'].isin(top_colors)]

plt.figure(figsize=(14, 7))
sns.barplot(data=df_colors, x='Engine', y='Price ($)', hue='Color', palette='muted')
plt.title('Car Price by Engine Type and Top Colors')
plt.xlabel('Engine Type')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.legend(title='Car Color')
plt.tight_layout()
plt.show()
```

```python
#Q12. Is there any seasonal trend in car sales based on the date of sale?

# Step 1: Convert Date column to pandas datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Step 2: Extract Month and Year from the Date column for aggregation
df['Month_Year'] = df['Date'].dt.to_period('M')

# Step 3: Group by Month_Year to count total sales (number of transactions)
monthly_sales = df.groupby('Month_Year').size().reset_index()
monthly_sales.columns = ['Month_Year', 'Sales_Count']

# Convert Month_Year back to timestamp format for plotting
monthly_sales['Month_Year'] = monthly_sales['Month_Year'].dt.to_timestamp()

print("--- Monthly Sales Trend ---")
print(monthly_sales)

# Step 4: Plot a line chart to visualize trends or seasonality
plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly_sales, x='Month_Year', y='Sales_Count', marker='o', color='red', linewidth=2)
plt.title('Monthly Trend in Car Sales')
plt.xlabel('Date of Sale (Months)')
plt.ylabel('Total Cars Sold')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q13. How does the car price distribution change when considering different combinations of body style and transmission type?

# Create a grouped box plot (using hue) to show the combinations
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Body Style', y='Price ($)', hue='Transmission', palette='pastel')
plt.title('Car Price Distribution by Body Style and Transmission Type')
plt.xlabel('Body Style')
plt.ylabel('Price ($)')
plt.legend(title='Transmission')
plt.show()
```

```python
#Q14. What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?

# Step 1: Prepare the dataset with continuous variables
corr_data = df[['Price ($)', 'Annual Income']]

# Step 2: Calculate the Pearson correlation matrix
corr_matrix = corr_data.corr()
print("--- Correlation Matrix ---")
print(corr_matrix)

# Step 3: Visualize the correlation matrix with a simple heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
#Q15. How does the average car price vary across different car models and engine types?

# Step 1: Group data by both Model and Engine to find the average price
model_engine_prices = df.groupby(['Model', 'Engine'])['Price ($)'].mean().reset_index()

# Step 2: Display the top 10 most expensive combinations
model_engine_prices = model_engine_prices.sort_values(by='Price ($)', ascending=False)
print("--- Top 10 Most Expensive Model & Engine Combinations ---")
print(model_engine_prices.head(10))

# Step 3: Create a heatmap pivot table using the top 10 models for clean visualization
top_10_models = df['Model'].value_counts().head(10).index
df_top_10 = df[df['Model'].isin(top_10_models)]

pivot_table = df_top_10.pivot_table(index='Model', columns='Engine', values='Price ($)', aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='YlGnBu', cbar_kws={'label': 'Average Price ($)'})
plt.title('Average Car Price Matrix by Top 10 Models and Engine Types')
plt.xlabel('Engine Type')
plt.ylabel('Car Model')
plt.tight_layout()
plt.show()
```
